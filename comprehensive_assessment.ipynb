### ข้อ 3: การใช้งาน Dictionary (Student Records)

ให้เขียนโปรแกรมเพื่อจัดการข้อมูลนักศึกษา:
1. สร้าง dictionary ของนักศึกษา 3 คน โดยแต่ละคนมี รหัส, ชื่อ, และ คะแนน
2. เพิ่มนักศึกษาคนที่ 4 ลงใน dictionary
3. อัปเดตคะแนนของนักศึกษาคนใดคนหนึ่ง
4. แสดงข้อมูลนักศึกษาทั้งหมด (ให้ใช้ loop)
5. คำนวณคะแนนเฉลี่ยของนักศึกษาทั้งหมด

**Expected Output:**
```
นักศึกษาทั้งหมด: 4 คน
คะแนนเฉลี่ย: XX.XX

[ข้อมูลแต่ละคน]
```

````python
# พิมพ์โค้ดคำตอบที่นี่
# ข้อมูลนักศึกษาเริ่มต้น 3 คน
students = {
    "6601001": {"name": "สมชาย", "score": 78},
    "6601002": {"name": "สมหญิง", "score": 85},
    "6601003": {"name": "สมหวัง", "score": 92}
}
students["6601004"] = {"name": "สมศรี", "score": 88}
students["6601001"]["score"] = 80
total_score = 0
print("ข้อมูลนักศึกษา:")
for sid, info in students.items():
    print("รหัส:", sid, "ชื่อ:", info["name"], "คะแนน:", info["score"])
    total_score += info["score"]
average = total_score / len(students)
print("\nนักศึกษาทั้งหมด:", len(students), "คน")
print("คะแนนเฉลี่ย:", format(average, ".2f"))
students = {
    "6601001": {"name": "สมชาย", "score": 78},
    "6601002": {"name": "สมหญิง", "score": 85},
    "6601003": {"name": "สมหวัง", "score": 92}
}
````

### ข้อ 4: นิพจน์ทางคณิตศาสตร์ (Mathematical Expressions)

ให้เขียนโปรแกรมแปลงนิพจน์ทางคณิตศาสตร์ต่อไปนี้เป็นโค้ด Python และแสดงผลลัพธ์:

1. **พื้นที่วงกลม:** A = πr² โดยให้ r = 5
2. **นิพจน์เศษส่วน:** (15 + 25) × 3 ÷ 2⁴
3. **รากที่ 3:** ∛64 (cube root of 64)
4. **ระยะทาง:** d = √((x₂-x₁)² + (y₂-y₁)²) โดยให้ (x₁,y₁) = (3,4) และ (x₂,y₂) = (7,1)

**หมายเหตุ:** ใช้ `math` module สำหรับ pi และ sqrt

**Expected Output:**
```
1. พื้นที่วงกลม = 78.50
2. ผลลัพธ์นิพจน์ = 7.50
3. รากที่ 3 ของ 64 = 4.00
4. ระยะทาง = 5.00
```

````python
import math

# พิมพ์โค้ดคำตอบที่นี่
import math
r = 5
area_circle = math.pi * r ** 2
print("1. พื้นที่วงกลม =", format(area_circle, ".2f"))
expression = (15 + 25) * 3 / (2 ** 4)
print("2. ผลลัพธ์นิพจน์ =", format(expression, ".2f"))
cube_root = 64 ** (1/3)
print("3. รากที่ 3 ของ 64 =", format(cube_root, ".2f"))
x1, y1 = 3, 4
x2, y2 = 7, 1
distance = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
print("4. ระยะทาง =", format(distance, ".2f"))
````

### ข้อ 5: ตัวดำเนินการเปรียบเทียบและเงื่อนไข (Comparison & Logic)

ให้เขียนโปรแกรมเพื่อ:
1. รับคะแนนสอบของนักศึกษา 3 วิชา
2. คำนวณคะแนนเฉลี่ย
3. ตรวจสอบผ่านหรือไม่ (เงื่อนไข: คะแนนเฉลี่ย >= 50 **และ** ทุกวิชา >= 40)
4. กำหนดเกรด โดยใช้เงื่อนไขต่อไปนี้:
   - A: คะแนน >= 80
   - B: คะแนน >= 70
   - C: คะแนน >= 60
   - D: คะแนน >= 50
   - F: คะแนน < 50

**ตัวอย่างการทำงาน:**
```python
# ให้รับค่าจากผู้ใช้หรือกำหนดค่าเอง
score1, score2, score3 = 75, 65, 80
```

**Expected Output (สำหรับตัวอย่างด้านบน):**
```
คะแนนวิชาที่ 1: 75
คะแนนวิชาที่ 2: 65
คะแนนวิชาที่ 3: 80
คะแนนเฉลี่ย: 73.33
ผลการสอบ: ผ่าน
เกรด: B
```

````python
# พิมพ์โค้ดคำตอบที่นี่
# กำหนดคะแนน (หรือจะใช้ input ก็ได้)
score1, score2, score3 = 75, 65, 80
print("คะแนนวิชาที่ 1:", score1)
print("คะแนนวิชาที่ 2:", score2)
print("คะแนนวิชาที่ 3:", score3)
average = (score1 + score2 + score3) / 3
print("คะแนนเฉลี่ย:", round(average, 2))
if average >= 50 and score1 >= 40 and score2 >= 40 and score3 >= 40:
    result = "ผ่าน"
else:
    result = "ไม่ผ่าน"
print("ผลการสอบ:", result)
if average >= 80:
    grade = "A"
elif average >= 70:
    grade = "B"
elif average >= 60:
    grade = "C"
elif average >= 50:
    grade = "D"
else:
    grade = "F"

print("เกรด:", grade)

score1, score2, score3 = 75, 65, 80
````

### ข้อ 6: การวนซ้ำและการกรองข้อมูล (Loop with Filtering)

ให้เขียนโปรแกรมที่แสดงตัวเลข 1 ถึง 50 โดยใช้ loop และจัดการดังนี้:
1. ถ้าตัวเลขหารด้วย 3 ลงตัว ให้ข้ามไป (ใช้ continue)
2. ถ้าตัวเลขหารด้วย 7 ลงตัว ให้แสดง "LUCKY" แทนตัวเลข
3. กรณีอื่น ให้แสดงตัวเลขตามปกติ
4. ถ้าเจอตัวเลขที่หารด้วย 37 ลงตัว ให้หยุดการวนซ้ำ (ใช้ break)

**Expected Output (ตัวอักษรแรกๆ):**
```
1 2 4 5 LUCKY 8 10 11 13 LUCKY 16 17 19 20 22 23 25 26 LUCKY 29 31 32 34 35 ...
```

````python
# พิมพ์โค้ดคำตอบที่นี่
for i in range(1, 51):
    if i % 37 == 0:
        break
    if i % 3 == 0:
        continue
    if i % 7 == 0:
        print("LUCKY", end=" ")
    else:
        print(i, end=" ")
````

1. รับ list ของตัวเลขเป็น parameter
2. คำนวณและ return dictionary ที่มี keys ต่อไปนี้:

### ข้อ 9: ฟังก์ชันตรวจสอบรหัสผ่าน (Password Validation Function)

ให้เขียนฟังก์ชัน `check_password(password)` ที่ตรวจสอบความปลอดภัยของรหัสผ่าน:

**เงื่อนไขการตรวจสอบ:**
1. ความยาวต้องอย่างน้อย 8 ตัวอักษร
2. ต้องมีตัวเลข (0-9) อย่างน้อย 1 ตัว
3. ต้องมีตัวอักษรพิมพ์ใหญ่ (A-Z) อย่างน้อย 1 ตัว

**ฟังก์ชันจะ return:**
- `True` ถ้า password ปลอดภัย
- `False` ถ้า password ไม่ปลอดภัย พร้อมกับข้อความบอกเหตุผล

**ตัวอย่างการใช้:**
```python
check_password("Hello123")     # True, "รหัสผ่านปลอดภัย"
check_password("hello")         # False
check_password("Hello12")       # False
check_password("HELLO123")      # True
```

````python
# พิมพ์โค้ดคำตอบที่นี่
def check_password(password):
    if len(password) < 8:
        return False, "รหัสผ่านต้องมีความยาวอย่างน้อย 8 ตัวอักษร"

    if not any(char.isdigit() for char in password):
        return False, "รหัสผ่านต้องมีตัวเลขอย่างน้อย 1 ตัว"

    if not any(char.isupper() for char in password):
        return False, "รหัสผ่านต้องมีตัวอักษรพิมพ์ใหญ่อย่างน้อย 1 ตัว"

    return True, "รหัสผ่านปลอดภัย"

# ให้สร้างฟังก์ชัน check_password แล้วทดสอบด้วยตัวอย่างต่างๆ

# test_passwords = ["Hello123", "hello", "Hello12", "HELLO123"]
# for pwd in test_passwords:
#     check_password(pwd)

````

### ข้อ 10: โปรแกรมจัดการร้านค้า (Inventory Management System)

ให้เขียนโปรแกรมจัดการคลังสินค้าที่มี dictionary เก็บข้อมูลสินค้า และฟังก์ชันต่างๆ ดังนี้:

**โครงสร้างข้อมูล:**
- Dictionary ชื่อ `inventory` มี structure: `{"ชื่อสินค้า": {"price": ราคา, "quantity": จำนวน}}`

**ฟังก์ชันที่ต้องสร้าง:**
1. `add_product(name, price, quantity)` - เพิ่มสินค้าใหม่ไปในคลังสินค้า
2. `sell_product(name, quantity)` - ขายสินค้า (ลดจำนวน) โดยต้องตรวจสอบว่ามีจำนวนพอหรือไม่
3. `show_inventory()` - แสดงรายการสินค้าทั้งหมด พร้อม ราคา จำนวน มูลค่า
4. `total_value()` - คำนวณมูลค่าสินค้าทั้งหมด (ราคา × จำนวน)

**ตัวอย่างการใช้:**
```python
inventory = {}

add_product("นม", 25, 10)
add_product("ขนมปัง", 15, 20)
add_product("ไข่", 30, 5)

sell_product("นม", 3)      # ขายนม 3 ชิ้น

show_inventory()
print(f"มูลค่าสินค้าทั้งหมด: {total_value()} บาท")
```

**Expected Output:**
```
===== รายการสินค้า ====="
1. นม
   ราคา: 25 บาท/ชิ้น
   คงเหลือ: 7 ชิ้น
   มูลค่า: 175 บาท

2. ขนมปัง
   ราคา: 15 บาท/ชิ้น
   คงเหลือ: 20 ชิ้น
   มูลค่า: 300 บาท

3. ไข่
   ราคา: 30 บาท/ชิ้น
   คงเหลือ: 5 ชิ้น
   มูลค่า: 150 บาท

========================
มูลค่าสินค้าทั้งหมด: 625 บาท
```

````python
# พิมพ์โค้ดคำตอบที่นี่
# พิมพ์โค้ดคำตอบที่นี่

# สร้าง dictionary สำหรับเก็บข้อมูลสินค้า
inventory = {}

def add_product(name, price, quantity):
    inventory[name] = {
        "price": price,
        "quantity": quantity
    }

def sell_product(name, quantity):
    if name in inventory:
        if inventory[name]["quantity"] >= quantity:
            inventory[name]["quantity"] -= quantity
        else:
            print(f"สินค้า {name} มีไม่พอ")
    else:
        print(f"ไม่พบสินค้า {name}")

def show_inventory():
    print("===== รายการสินค้า =====")
    count = 1
    for name, info in inventory.items():
        price = info["price"]
        quantity = info["quantity"]
        value = price * quantity

        print(f"{count}. {name}")
        print(f"ราคา: {price} บาท/ชิ้น")
        print(f"คงเหลือ: {quantity} ชิ้น")
        print(f"มูลค่า: {value} บาท\n")
        count += 1
    print("========================")

def total_value():
    total = 0
    for info in inventory.values():
        total += info["price"] * info["quantity"]
    return total

add_product("นม", 25, 10)
add_product("ขนมปัง", 15, 20)
add_product("ไข่", 30, 5)

sell_product("นม", 3)

show_inventory()
print(f"มูลค่าสินค้าทั้งหมด: {total_value()} บาท")
เขียนถึง ภาระกิจลอกการบ้านเพื่อน

inventory = {}

````
