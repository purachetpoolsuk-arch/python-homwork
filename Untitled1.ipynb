```python
class Student:
    def __init__(self, student_id, citizen_id, first_name, last_name, age, gender):
        self.student_id = student_id          # รหัสนักศึกษา
        self.citizen_id = citizen_id          # เลขบัตรประชาชน 13 หลัก
        self.first_name = first_name          # ชื่อ
        self.last_name = last_name            # นามสกุล
        self.age = age                        # อายุ
        self.gender = gender                  # เพศ

    def show_info(self):
        print("ข้อมูลนักศึกษา")
        print(f"รหัสนักศึกษา: {self.student_id}")
        print(f"เลขบัตรประชาชน: {self.citizen_id}")
        print(f"ชื่อ-นามสกุล: {self.first_name} {self.last_name}")
        print(f"อายุ: {self.age} ปี")
        print(f"เพศ: {self.gender}")
```

```python

student1 = Student(
    student_id="6812732114",
    citizen_id="1209702255691",
    first_name="ปุรเชษฐ์",
    last_name="พูลสุข",
    age=20,
    gender="ชาย"
)

student1.show_info()
```
